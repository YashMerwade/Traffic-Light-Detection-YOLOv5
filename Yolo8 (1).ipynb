```python
import os, shutil

base = "/content/S2TLD/S2TLD（720x1280）"
merged_dir = "/content/S2TLD_merged"
os.makedirs(merged_dir + "/Annotations", exist_ok=True)
os.makedirs(merged_dir + "/JPEGImages", exist_ok=True)

for subdir in ["normal_1", "normal_2"]:
    ann_src = os.path.join(base, subdir, "Annotations")
    img_src = os.path.join(base, subdir, "JPEGImages")

    for f in os.listdir(ann_src):
        shutil.copy(os.path.join(ann_src, f), os.path.join(merged_dir, "Annotations", f))
    for f in os.listdir(img_src):
        shutil.copy(os.path.join(img_src, f), os.path.join(merged_dir, "JPEGImages", f))

print("✅ All merged into", merged_dir)
```

```python
import os
import glob
import random
from sklearn.model_selection import train_test_split
from xml.etree import ElementTree as ET

# Define paths
merged_ann_dir = "/content/S2TLD_merged/Annotations"
merged_img_dir = "/content/S2TLD_merged/JPEGImages"
output_dir = "/content/S2TLD-YOLO"
os.makedirs(output_dir, exist_ok=True)

# Define YOLO folder structure
for split in ["train", "val", "test"]:
    os.makedirs(f"{output_dir}/images/{split}", exist_ok=True)
    os.makedirs(f"{output_dir}/labels/{split}", exist_ok=True)

# Define classes
class_map = {"red": 0, "yellow": 1, "green": 2, "False": 3, "wait": 4}

# List all XML files
xml_files = glob.glob(os.path.join(merged_ann_dir, "*.xml"))

# Shuffle and split
trainval, test = train_test_split(xml_files, test_size=0.15, random_state=42)
train, val = train_test_split(trainval, test_size=0.175, random_state=42)  # 68/15/17 split

def convert_to_yolo(xml_path, out_txt_path, img_w, img_h):
    tree = ET.parse(xml_path)
    root = tree.getroot()
    yolo_lines = []

    for obj in root.findall("object"):
        label = obj.find("name").text.strip()
        if label not in class_map:
            continue

        bbox = obj.find("bndbox")
        xmin = int(float(bbox.find("xmin").text))
        xmax = int(float(bbox.find("xmax").text))
        ymin = int(float(bbox.find("ymin").text))
        ymax = int(float(bbox.find("ymax").text))

        x_center = (xmin + xmax) / 2 / img_w
        y_center = (ymin + ymax) / 2 / img_h
        width = (xmax - xmin) / img_w
        height = (ymax - ymin) / img_h

        yolo_lines.append(f"{class_map[label]} {x_center:.6f} {y_center:.6f} {width:.6f} {height:.6f}")

    if yolo_lines:
        with open(out_txt_path, "w") as f:
            f.write("\n".join(yolo_lines))

def process_split(split_list, split_name):
    for xml in split_list:
        img_name = os.path.basename(xml).replace(".xml", ".jpg")
        img_path = os.path.join(merged_img_dir, img_name)
        target_img = os.path.join(output_dir, "images", split_name, img_name)
        target_txt = os.path.join(output_dir, "labels", split_name, img_name.replace(".jpg", ".txt"))

        if not os.path.exists(img_path):
            continue

        # Copy image
        shutil.copy(img_path, target_img)

        # Use fixed size 1280x720 (from dataset name)
        convert_to_yolo(xml, target_txt, img_w=1280, img_h=720)

process_split(train, "train")
process_split(val, "val")
process_split(test, "test")

print("✅ Done! YOLO formatted dataset created at:", output_dir)
```

```python
# Re-import necessary stuff
from pathlib import Path
import random

img_dir = Path("/content/S2TLD_yolo/images")
lbl_dir = Path("/content/S2TLD_yolo/labels")

# Create subdirs
for split in ["train", "val", "test"]:
    (img_dir / split).mkdir(parents=True, exist_ok=True)
    (lbl_dir / split).mkdir(parents=True, exist_ok=True)

# List all image files
img_files = list(img_dir.glob("*.jpg")) + list(img_dir.glob("*.jpeg")) + list(img_dir.glob("*.png"))
random.shuffle(img_files)

n = len(img_files)
n_train = int(n * 0.7)
n_val = int(n * 0.15)
n_test = n - n_train - n_val

train_files = img_files[:n_train]
val_files = img_files[n_train:n_train + n_val]
test_files = img_files[n_train + n_val:]

def move_pairs(files, target_img_dir, target_lbl_dir):
    for img_path in files:
        lbl_path = lbl_dir / (img_path.stem + ".txt")
        if lbl_path.exists():
            shutil.move(str(img_path), str(target_img_dir / img_path.name))
            shutil.move(str(lbl_path), str(target_lbl_dir / lbl_path.name))

move_pairs(train_files, img_dir / "train", lbl_dir / "train")
move_pairs(val_files, img_dir / "val", lbl_dir / "val")
move_pairs(test_files, img_dir / "test", lbl_dir / "test")

print(f"✅ Split complete: {len(train_files)} train, {len(val_files)} val, {len(test_files)} test")
```

```python
import cv2
from ultralytics import YOLO

model = YOLO('/content/yolo_train_runs/weights/best.pt')

video_path = '/content/drive/MyDrive/Video Traffic Light2.mp4'
cap = cv2.VideoCapture(video_path)

# Define video writer
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('/content/output_video.mp4', fourcc, 30.0, (int(cap.get(3)), int(cap.get(4))))

results = model(source=video_path, stream=True, conf=0.4, iou=0.4, imgsz=960, classes=[0, 1, 2], device='cpu')

for r in results:
    frame = r.orig_img
    annotated_frame = r.plot()  # Draw detections
    out.write(annotated_frame)  # Save frame

cap.release()
out.release()
print("✅ Output saved to /content/output_video.mp4")
```
