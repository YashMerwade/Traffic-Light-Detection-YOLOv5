```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Simulated confusion matrix data (example for 4 classes)
conf_matrix_raw = np.array([
    [50, 2, 1, 0],
    [4, 45, 0, 1],
    [2, 0, 47, 3],
    [1, 1, 2, 46]
])

# Normalize by column (like YOLO's default behavior)
conf_matrix_norm = conf_matrix_raw / conf_matrix_raw.sum(axis=0, keepdims=True)

# Class labels
class_names = ['Class A', 'Class B', 'Class C', 'Class D']

# Plot raw confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_raw, annot=True, fmt="d", cmap="Blues",
            xticklabels=class_names, yticklabels=class_names)
plt.title("Raw Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.tight_layout()
plt.savefig("raw_confusion_matrix.png")
plt.show()

# Plot normalized confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_norm, annot=True, fmt=".2f", cmap="Greens",
            xticklabels=class_names, yticklabels=class_names)
plt.title("Normalized Confusion Matrix (by Column)")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.tight_layout()
plt.savefig("normalized_confusion_matrix.png")
plt.show()
```
